```python
region_sales = df.groupby('Region')['Sales'].sum()
region_sales
```

```python
region_sales.plot(kind='bar', title='Total Sales by Region')
plt.ylabel('Sales')
plt.xlable('Region')
plt.show
```

```python
category_sales = df.groupby('Category')['Sales'].sum()
category_sales
```

```python
category_sales.plot(kind='bar', title='Total Sales by Category')
plt.ylabel('Sales')
plt.xlable('Category')
plt.show()
```

```python
product_profit = df.groupby('Product')['Profit'].sum().sort_values(ascending=False)
product_profit
```

```python
product_profit.plot(kind='bar', title='Total Product by Profit')
plt.ylable('Profit')
plt.xlable('Product')
plt.show()
```
